```python
def is_prime(x):
    if x == 2 or x == 1:
        return True
    else:
        for divisor in range (2,x):
            if x % divisor == 0:
                return False
            if divisor == (x-1):
                return True 
```

```python
def is_even(x):
    if x % 2 == 0 :
        return True
    else:
        return False
```

```python
def even_prime_list_creation(even,prime):
    return {"prime":prime,"even":even}
```

# Proper Solution

```python
def is_prime(number,divisor):
    if number == 2 or number == 1:
        return True
    else:
        if divisor == number:
            return True
        if number % divisor == 0:
            return False
        bool_value = is_prime(number,divisor+1)
    return bool_value
```

```python
def even_prime_list_creation(integers,even,prime):
    if len(integers) == 0:
        return even,prime
    if integers[0] % 2 == 0:
        even.append(integers[0])
    else:
        if is_prime(integers[0],2) == True:
            prime.append(integers[0])
    integers.pop(0)
    even_list,prime_list = even_prime_list_creation(integers,even,prime)
    return even_list,prime_list
```

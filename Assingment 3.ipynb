```python
import numpy as np
integers = np.random.randint(low=0, high=2000, size=(1000,))
```

```python
integers
```

```python
def is_prime(x):
    if x == 2 or x == 1:
        return True
    else:
        for divisor in range (2,x):
            if x % divisor == 0:
                return False
            if divisor == (x-1):
                return True 
```

```python
def is_even(x):
    if x % 2 == 0 :
        return True
    else:
        return False
```

```python
def even_prime_list_creation(even,prime):
    return {"prime":prime,"even":even}
```

```python
prime = []
even = []
for x in integers:
    if is_prime(x)==True:
        prime.append(x)
    else:
        
        if is_even(x)==True:
            even.append(x)
```

```python
even_prime_list_creation(even,prime)
```
